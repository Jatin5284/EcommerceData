```python
import mysql.connector as my

con = my.connect(
    host = "localhost",
    user = "root",
    password = "",
    database = "ecommerce"
)

cursor = con.cursor()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

# Query data from the 'customer' table
cursor.execute('SELECT * FROM customer')
#After fetching data from the database we are storing it into Pandas DataFrame
customer = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Query data from the 'product' table
cursor.execute('SELECT * FROM product')
product = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Query data from the 'order_details' table
cursor.execute('SELECT * FROM order_details')
order = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])
```

```python
# Aggregate the number of customers by city
city_counts = customer['city'].value_counts()
print(city_counts)
```

```python
# Create and display a bar graph
plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar', color='lightblue')
plt.title('Number of Customers  City-Wise')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count orders per customer
order_counts = order['customer_id'].value_counts()

# Convert to DataFrame for better readability
order_counts_df = order_counts.reset_index()
order_counts_df.columns = ['customer_id', 'number_of_orders']

# Merge with customer data to get names
customer_orders_df = pd.merge(order_counts_df, customer, on='customer_id')

# Sort by number of orders in descending order
customer_orders_df = customer_orders_df.sort_values(by='number_of_orders', ascending=False)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(customer_orders_df['name'], customer_orders_df['number_of_orders'], color='lightblue')
plt.title('Most Frequent Customers by Number of Orders')
plt.xlabel('Customer Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Count products per category
category_counts = product['category'].value_counts()

# Convert to DataFrame for better readability
category_counts_df = category_counts.reset_index()
category_counts_df.columns = ['Category', 'Number of Products']

# Plotting the bar chart
plt.figure(figsize=(7, 4))
plt.bar(category_counts_df['Category'], category_counts_df['Number of Products'], color='lightblue')
plt.title('Distribution of Products')
plt.xlabel('Product ID')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Count the number of products per sub-category
sub_category_counts = product['sub_category'].value_counts()

# Convert to DataFrame for better readability
sub_category_counts_df = sub_category_counts.reset_index()
sub_category_counts_df.columns = ['Sub-Category', 'Number of Products']

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(sub_category_counts_df['Sub-Category'], sub_category_counts_df['Number of Products'], color='blue')
plt.title('Distribution of Products Across Sub-Categories')
plt.xlabel('Sub-Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Define the low stock threshold
low_stock_threshold = 30

# Filter products with stock levels below or equal to the threshold
low_stock_products = product[product['stock'] <= low_stock_threshold]

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(low_stock_products['product_name'], low_stock_products['stock'], color='lightblue')
plt.title('Products with Low Stock Levels')
plt.xlabel('Product Name')
plt.ylabel('Stock Level')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Calculate statistics
average_price = product['selling_price'].mean()
max_price = product['selling_price'].max()
min_price = product['selling_price'].min()

# Prepare data for the bar graph
statistics = {
    'Average Price': average_price,
    'Maximum Price': max_price,
    'Minimum Price': min_price
}

# Create and display a bar graph
plt.figure(figsize=(7, 4))
plt.bar(statistics.keys(), statistics.values(), color=['orange', 'green', 'blue'])
plt.title('Average , Maximum , Minimum Prices of Product')
plt.xlabel('Price Type')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Query data from the 'product' table to get product names
cursor.execute('SELECT product_id, product_name FROM product')

product_names = pd.DataFrame(cursor.fetchall(), columns=['product_id', 'product_name'])

# Merge the order data with product names
order = order.merge(product_names, on='product_id')

# Calculate total quantity ordered by each product
product_quantity = order.groupby('product_name').agg({'quantity': 'sum'}).reset_index()

# Sort the products by total quantity in descending order and get the top 10
top_10_products = product_quantity.sort_values(by='quantity', ascending=False).head(10)

# Display the top 10 products
print(top_10_products)

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(top_10_products['product_name'], top_10_products['quantity'], color='cornflowerblue')
plt.title('Total Number of Orders Product-wise with Product Names')
plt.xlabel('Product Name')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()

# Display the chart
plt.show()


```

```python
status_counts = order['order_status'].value_counts()
print(status_counts)

# Plotting
status_counts.plot(kind='bar', figsize=(10, 7))
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.show()
```

```python
popular_products = order.groupby('product_name').agg({'quantity': 'sum'}).reset_index()
popular_products = popular_products.sort_values(by='quantity', ascending=False).head(10)
print(popular_products)

# Plotting
popular_products.plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Most Popular Products by Quantity')
plt.xlabel('Total Order Quantity')
plt.ylabel('Product Name')

plt.yticks(range(len(popular_products)), popular_products['product_name'])

plt.show()
```

```python
revenue_by_product = order.groupby('product_name').agg({'total_price': 'sum'}).reset_index()
print(revenue_by_product)

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(revenue_by_product['product_name'], revenue_by_product['total_price'])
plt.title('Top Products by Revenuie')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
revenue_by_category = order.merge(product[['product_id', 'category']], on='product_id')
revenue_by_category = revenue_by_category.groupby('category').agg({'total_price': 'sum'}).reset_index()
revenue_by_category['percentage'] = (revenue_by_category['total_price'] / revenue_by_category['total_price'].sum()) * 100
print(revenue_by_category,"\n")

# Plotting
plt.figure(figsize=(10, 7))
plt.pie(revenue_by_category['percentage'], labels=revenue_by_category['category'], autopct='%1.1f%%')
plt.title('Revenue Percentage by Product Category')
plt.show()
```

```python

# Query to get total revenue from all orders
cursor.execute('''
    SELECT SUM(total_price) AS total_revenue
    FROM order_details
''')
# Fetch the result
result = cursor.fetchone()

# Extract total revenue from the result
total_revenue = result[0]

# Create a DataFrame for plotting
data = {'Total Revenue': [total_revenue]}
df = pd.DataFrame(data)

# Create and display a bar graph
plt.figure(figsize=(8, 6))
plt.bar(df.index, df['Total Revenue'], color='skyblue')
plt.title('Total Revenue Generated from All Orders')
plt.xlabel('Metrics')
plt.ylabel('Amount ($)')
plt.xticks(df.index, ['Total Revenue'])
plt.tight_layout()
plt.show()
```

```python

# Query data to get product details with original and selling prices
cursor.execute('''
    SELECT product_id, product_name, original_price, selling_price
    FROM product
''')
product_data = pd.DataFrame(cursor.fetchall(), columns=['product_id', 'product_name', 'original_price', 'selling_price'])

# Calculate profit for each product
product_data['profit'] = product_data['selling_price'] - product_data['original_price']

# Sort products by profit in descending order
product_data = product_data.sort_values(by='profit', ascending=False)

# Display the profit data
print(product_data[['product_name', 'profit']])

# Plotting the bar chart
plt.figure(figsize=(12, 8))
plt.bar(product_data['product_name'], product_data['profit'])
plt.title('Most Profitable Products')
plt.xlabel('Product Name')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Display the chart
plt.show()

```

```python
# Query data to get total quantity ordered per product
cursor.execute('''
    SELECT p.product_name, SUM(od.quantity) AS total_quantity
    FROM order_details od
    JOIN product p ON od.product_id = p.product_id
    GROUP BY p.product_name
    ORDER BY total_quantity DESC
''')
quantity_data = pd.DataFrame(cursor.fetchall(), columns=['product_name', 'total_quantity'])

# Identify products with highest and lowest order quantities
highest_quantity_product = quantity_data.iloc[0]
lowest_quantity_product = quantity_data.iloc[-1]

# Prepare data for the bar chart
chart_data = pd.DataFrame({
    'product_name': [highest_quantity_product['product_name'], lowest_quantity_product['product_name']],
    'total_quantity': [highest_quantity_product['total_quantity'], lowest_quantity_product['total_quantity']]
})

# Display the quantities data
print(chart_data)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(chart_data['product_name'], chart_data['total_quantity'], color=['red', 'green'])
plt.title('Total Quantity of Highest and Lowest Order Products')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()

# Display the chart
plt.show()

```

```python
# Merge order_details with customer to get customer names
merged_data = pd.merge(order, customer, on='customer_id', how='left')

# Calculate the total quantity per customer
total_quantity_per_customer = merged_data.groupby('customer_id')['quantity'].sum()

# Find the customer IDs with the highest and lowest total quantity
highest = total_quantity_per_customer.idxmax()
lowest = total_quantity_per_customer.idxmin()

# Get the names of the highest and lowest customers
highest_name = customer.set_index('customer_id').loc[highest, 'name']
lowest_name = customer.set_index('customer_id').loc[lowest, 'name']

# Data for plotting
customers = [highest_name, lowest_name]
quantities = [total_quantity_per_customer[highest], total_quantity_per_customer[lowest]]

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(customers, quantities)
plt.title('Top Order Quantity by Customer (Highest and Lowest)')
plt.xlabel('Customer Name')
plt.ylabel('Total Order Quantity')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Query data to get counts of orders by payment mode
cursor.execute('''
    SELECT payment_mode, COUNT(*) AS order_count
    FROM order_details
    GROUP BY payment_mode
    ORDER BY order_count DESC
''')
payment_mode_data = pd.DataFrame(cursor.fetchall(), columns=['payment_mode', 'order_count'])

# Display the payment mode data
print(payment_mode_data)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(payment_mode_data['payment_mode'], payment_mode_data['order_count'])
plt.title('Payment Mode Preferences')
plt.xlabel('Payment Mode')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')  

# Display the chart
plt.show()

```

```python
order['order_date'] = pd.to_datetime(order['order_date'])

order['year_month_date'] = order['order_date'].dt.to_period('D')

monthly_sales = order.groupby('year_month_date').agg({'total_price': 'sum'}).reset_index()

monthly_sales['year_month_date'] = monthly_sales['year_month_date'].dt.to_timestamp()

plt.figure(figsize=(12, 8))
plt.bar(monthly_sales['year_month_date'], monthly_sales['total_price'])
plt.title('Month-Wise Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45,ha='right')  # Rotate x labels for better readability

# Display the chart
plt.show()
```

```python
# Analyze the distribution of customers across cities
city_distribution = customer['city'].value_counts().reset_index()
city_distribution.columns = ['city', 'customer_count']

# Display the DataFrame
print(city_distribution)

# Create a bar plot for city distribution
plt.figure(figsize=(12, 8))
sns.barplot(x='customer_count', y='city', data=city_distribution, palette='viridis',hue='city')

# Add labels and title
plt.xlabel('Number of Customers')
plt.ylabel('City')
plt.title('Distribution of Customers Across Different Cities')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
merged_df = pd.merge(order, customer[['customer_id', 'city']], on='customer_id')
# merged_df = pd.merge(merged_df, product[['product_id', 'category']], on='product_id')
heatmap_data = merged_df.pivot_table(index='city', columns='product_name', values='quantity', aggfunc='sum',fill_value=0)
# print(pivot_df)
# Visualize the results using a heatmap

# sns.heatmap(pivot_df, annot=True, cmap='Blues')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='g')
plt.xlabel('Category')
plt.ylabel('City')
plt.title('Heatmap of Category Popularity by City')
plt.tight_layout()
plt.show()
```

```python
#  Product Performance:
#  Identify the best-selling products

merged = pd.merge(order, product[['product_id', 'product_name']], on='product_id')
# print(merged)
# Group data by product name and calculate total order quantity
best_selling_products = merged.groupby('product_name_x')['quantity'].sum().reset_index()
# Sort results in descending order
best_selling_products = best_selling_products.sort_values('quantity', ascending=False)

# Visualize the results
plt.figure(figsize=(10, 6))
plt.bar(best_selling_products['product_name_x'], best_selling_products['quantity'])
plt.title('Best-Selling Products')
plt.xlabel('Product Name ')
plt.ylabel('Total  Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.show()


```
